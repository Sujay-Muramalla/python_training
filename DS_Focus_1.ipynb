## Q1. If-Else: Positive/Negative/Zero
Write a program to check whether a number is positive, negative, or zero.
Input: an integer `n`
Output: one of: `Positive`, `Negative`, `Zero`.

```python
inp = int (input("enter number"))

if (inp > 0):
    print("positive")
elif (inp < 0):
    print ("negative")
else:
    print ("zero")
    
```
