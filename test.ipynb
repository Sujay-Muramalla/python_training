```python
def interleave(str1,str2):
    #str3 = ''.join(str(list (zip(str1,str2))))
    list1= list(str1)
    list2= list(str2)
    print (f"list1: {list1}")
    print (f"list2: {list2}")
    list3 = list(zip(list1,list2))
    print (f"list3: {list3}")
    str3 = ''.join (''.join(pair) for pair in list3)
    print (str3)
    return str3

print (interleave('hi','ha'))    # 'hhia'
print (interleave('aaa', 'zzz'))  # 'azazaz'
print (interleave('lzr','iad')) #  'lizard'
```
